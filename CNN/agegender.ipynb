```python
# Import required modules
import cv2 as cv
import math
import time
import argparse
```

```python

def getFaceBox(net, frame, conf_threshold=0.7):
    frameOpencvDnn = frame.copy()
    frameHeight = frameOpencvDnn.shape[0]
    frameWidth = frameOpencvDnn.shape[1]
    blob = cv.dnn.blobFromImage(frameOpencvDnn, 1.0, (300, 300), [104, 117, 123], True, False)

    net.setInput(blob)
    detections = net.forward()
    bboxes = []
    for i in range(detections.shape[2]):
        confidence = detections[0, 0, i, 2]
        if confidence > conf_threshold:
            x1 = int(detections[0, 0, i, 3] * frameWidth)
            y1 = int(detections[0, 0, i, 4] * frameHeight)
            x2 = int(detections[0, 0, i, 5] * frameWidth)
            y2 = int(detections[0, 0, i, 6] * frameHeight)
            bboxes.append([x1, y1, x2, y2])
            cv.rectangle(frameOpencvDnn, (x1, y1), (x2, y2), (0, 255, 0), int(round(frameHeight/150)), 8)
    return frameOpencvDnn, bboxes

```

```python


faceProto = "opencv_face_detector.pbtxt"
faceModel = "opencv_face_detector_uint8.pb"

ageProto = "age_deploy.prototxt"
ageModel = "age_net.caffemodel"

genderProto = "gender_deploy.prototxt"
genderModel = "gender_net.caffemodel"
```

```python

MODEL_MEAN_VALUES = (78.4263377603, 87.7689143744, 114.895847746)
ageList = ['(0-2)', '(4-6)', '(8-12)', '(15-20)', '(25-32)', '(38-43)', '(48-53)', '(60-100)']
genderList = ['Male', 'Female']

```

```python
# Load network
ageNet = cv.dnn.readNet(ageModel, ageProto)
genderNet = cv.dnn.readNet(genderModel, genderProto)
faceNet = cv.dnn.readNet(faceModel, faceProto)


```

```python

# Open a video file or an image file or a camera stream
cap = cv.VideoCapture('MLG5.mp4')
padding = 20
```

```python
g = 0
b = 0

ga1 =0
ga2 =0
ga3 =0
ga4 =0
ga5 =0
ga6 =0
ga7 =0
ga8 =0

ba1 =0
ba2 =0
ba3 =0
ba4 =0
ba5 =0
ba6 =0
ba7 =0
ba8 =0

gsum1 = 0
gsum2 = 0
gsum3 = 0
gsum4 = 0
gsum5 = 0
gsum6 = 0
gsum7 = 0
gsum8 = 0

bsum1 = 0
bsum2 = 0
bsum3 = 0
bsum4 = 0
bsum5 = 0
bsum6 = 0
bsum7 = 0
bsum8 = 0


while cv.waitKey(1) < 0:
    # Read frame
    t = time.time()
    hasFrame, frame = cap.read()
    if not hasFrame:
        cv.waitKey()
        break

    frameFace, bboxes = getFaceBox(faceNet, frame)
    if not bboxes:
        print("No face Detected, Checking next frame")
        continue

    for bbox in bboxes:
        # print(bbox)
        face = frame[max(0,bbox[1]-padding):min(bbox[3]+padding,frame.shape[0]-1),max(0,bbox[0]-padding):min(bbox[2]+padding, frame.shape[1]-1)]

        blob = cv.dnn.blobFromImage(face, 1.0, (227, 227), MODEL_MEAN_VALUES, swapRB=False)
        
        genderNet.setInput(blob)
        genderPreds = genderNet.forward()
        gender = genderList[genderPreds[0].argmax()]
        #print("Gender Output : {}".format(genderPreds))
        print("Gender : {}, conf = {:.3f}".format(gender, genderPreds[0].max()))
        
      
        
        ageNet.setInput(blob)
        agePreds = ageNet.forward()
        age = ageList[agePreds[0].argmax()]
        #print("Age Output : {}".format(agePreds))
        print("Age : {}, conf = {:.3f}".format(age, agePreds[0].max()))
        
        if(format(gender) == 'Female') :
            g = g+1
            
            if(format(age) == '(0-2)') :
                ga1 = ga1+1
                gsum1 = gsum1+agePreds[0].max()
            if(format(age) == '(4-6)') :
                ga2 = ga2+1
                gsum2 = gsum2+agePreds[0].max()
            if(format(age) == '(8-12)') :
                ga3 = ga3+1
                gsum3 = gsum3+agePreds[0].max()
            if(format(age) == '(15-20)') :
                ga4 = ga4+1
                gsum4 = gsum4+agePreds[0].max()
            if(format(age) == '(25-32)') :
                ga5 = ga5+1
                gsum5 = gsum5+agePreds[0].max()
            if(format(age) == '(38-43)') :
                ga6 = ga6+1
                gsum6 = gsum6+agePreds[0].max()
            if(format(age) == '(48-53)') :
                ga7 = ga7+1
                gsum7 = gsum7+agePreds[0].max()
            if(format(age) == '(60-100)') :
                ga8 = ga8+1      
                gsum8 = gsum8+agePreds[0].max()
            
            
        if(format(gender) == 'Male') :
            b = b+1
        
            if(format(age) == '(0-2)') :
                ba1 = ba1+1
                bsum1 = bsum1+agePreds[0].max()
            if(format(age) == '(4-6)') :
                ba2 = ba2+1
                bsum2 = bsum2+agePreds[0].max()
            if(format(age) == '(8-12)') :
                ba3 = ba3+1
                bsum3 = bsum3+agePreds[0].max()
            if(format(age) == '(15-20)') :
                ba4 = ba4+1
                bsum4 = bsum4+agePreds[0].max()
            if(format(age) == '(25-32)') :
                ba5 = ba5+1
                bsum5 = bsum5+agePreds[0].max()
            if(format(age) == '(38-43)') :
                ba6 = ba6+1
                bsum6 = bsum6+agePreds[0].max()
            if(format(age) == '(48-53)') :
                ba7 = ba7+1
                bsum7 = bsum7+agePreds[0].max()
            if(format(age) == '(60-100)') :
                ba8 = ba8+1      
                bsum8 = bsum8+agePreds[0].max()
        
        print("Female : {}".format(g))
        print(ga1, round(gsum1,2),"|", ga2, round(gsum2,2),"|", ga3, round(gsum3,2),"|", ga4, round(gsum4,2),"|", ga5, round(gsum5,2),"|", ga6, round(gsum6,2),"|", ga7, round(gsum7,2),"|", ga8, round(gsum8,2))   

        print("Male : {}".format(b))
        print(ba1, round(bsum1,2),"|", ba2, round(bsum2,2),"|", ba3, round(bsum3,2),"|", ba4, round(bsum4,2),"|", ba5, round(bsum5,2),"|", ba6, round(bsum6,2),"|", ba7, round(bsum7,2),"|", ba8, round(bsum8,2))   
        
        
        label = "{},{}".format(gender, age)
        cv.putText(frameFace, label, (bbox[0], bbox[1]-10), cv.FONT_HERSHEY_SIMPLEX, 0.8, (0, 255, 255), 2, cv.LINE_AA)
        cv.imshow("Age Gender Demo", frameFace)
        
        
        
        # cv.imwrite("../Users/chaejeong/Downloads/images/a.png", frameFace)
        # cv.imwrite("age-gender-out-{}".format(args.input),frameFace)
    #print("time : {:.3f}".format(time.time() - t))



```
